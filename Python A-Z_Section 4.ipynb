```python
# create first function 
# The function will be a wrapper for all of the code above

def myplot(playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c="Black", ls = "--", marker = "s", ms=5, label = name)

    plt.legend(loc = "upper left", bbox_to_anchor=(1.025,1.025))
    plt.xticks(list(range(0,10)),Seasons, rotation = "vertical")
    plt.show()
```
